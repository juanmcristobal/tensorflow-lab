# Test GPU and Machine Learning Libraries

This notebook verifies that the GPU is available and tests basic functionality of the installed libraries.

```python
# Check TensorFlow GPU availability
# https://www.tensorflow.org/install/source#gpu

import tensorflow as tf
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

print("TensorFlow Version:", tf.__version__)
print("Is GPU available:", tf.config.list_physical_devices('GPU'))
```

```python
# Simple TensorFlow GPU computation
with tf.device('/GPU:0'):
    a = tf.constant([[1.0, 2.0], [3.0, 4.0]])
    b = tf.constant([[5.0, 6.0], [7.0, 8.0]])
    c = tf.matmul(a, b)
    print("Matrix multiplication result:")
    print(c)
```

```python
# Test basic NumPy functionality
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])
print("NumPy Array:")
print(arr)
print("Array Mean:", np.mean(arr))
```

```python
# Test pandas functionality
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "City": ["New York", "San Francisco", "Los Angeles"]
}
df = pd.DataFrame(data)
print("Pandas DataFrame:")
print(df)
```

```python
# Test matplotlib visualization
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

plt.plot(x, y, label="Linear Growth")
plt.title("Matplotlib Test")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
# Test Seaborn visualization
import seaborn as sns

tips = sns.load_dataset("tips")
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
plt.title("Seaborn Test")
plt.show()
```
